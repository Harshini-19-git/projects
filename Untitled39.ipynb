```python
import google.generativeai as genai

def extract_spec_gemini(query, text):

    prompt = f"""
You are an automotive specification extractor.
Return vehicle specs in ONLY valid JSON.

Format output like:
[
  {{
    "component": "",
    "spec_type": "",
    "value": "",
    "unit": ""
  }}
]

If no measurement or specification found ‚Üí return [].

Query: {query}
Text:
{text}
"""

    model = genai.GenerativeModel("gemini-2.5-flash")
    response = model.generate_content(prompt)
    return response.text
```
